```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
DATA_DIR = r"D:\ml\face"
os.listdir(DATA_DIR)
```

```python
def load_images_labels(datadir , label = 0):
    images = []
    labels = []
    for i in os.listdir(datadir):
        img = cv2.imread(os.path.join(datadir,i))
        images.append(img)
        labels.append(label)
    images = np.array(images)
    labels = np.array(labels).reshape(-1,1)
    return images , labels
```

```python
datadir = os.path.join(DATA_DIR,'sayan')
my_images , my_labels = load_images_labels(datadir,label=0)
```

```python
datadir = os.path.join(DATA_DIR,'pratyay')
my_images2 , my_labels2 = load_images_labels(datadir , label = 1)
```

```python
x_train , y_train = np.vstack([my_images,my_images2]), np.vstack([my_labels,my_labels2])
```

```python
y_train.shape
```

```python
import pickle
with open('train_images1.p','wb') as f:
    pickle.dump(x_train , f)
with open('train_labels2.p','wb') as f:
    pickle.dump(y_train ,f)
```
